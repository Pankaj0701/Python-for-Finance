```python
import numpy as np
import pandas as pd
import yfinance as yf
import pandas_datareader
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
PG = yf.download('PG', start = '1995-1-1')
```

```python
PG.head()
```

```python
PG['log_return'] = np.log(PG['Close'] / PG['Close'].shift(1))
print(PG['log_return'])
```

```python
PG['log_return'].plot(figsize=(8,5))
```

```python
log_return_d = (PG['log_return'].mean()).item() 
log_return_d
```

```python
log_return_a = (PG['log_return'].mean()*250).item()
log_return_a
```

```python
print(str(round(log_return_a * 100, 3)) + '%')
```
